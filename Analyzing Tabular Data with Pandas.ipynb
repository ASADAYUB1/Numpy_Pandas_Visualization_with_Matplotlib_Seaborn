### Analyzing Data from Data Frames 

```python
total_cases = covid_df.new_cases.sum()
total_deaths = covid_df.new_deaths.sum()
print('The total number of cases report are {} and the number total of reported deaths are {}.' .format(int(total_cases), int(total_deaths)))
```

```python
# overall death ratio 
death_rate = (total_deaths / total_cases)*100
print('The death raio in pakistan during the Covid was {:.2f}%.'.format(death_rate))
```

```python
initail_tests = 935310
total_tests = initail_tests + int(covid_df.new_tests.sum() )
```

```python
total_tests
```

```python
positive_rate = total_cases / total_tests
print('The {:.2f}% of test in pakistan leds to positive.' .format(positive_rate))
```

- `sort_values` - Sorting the rows of a data frame using column values

```python
# Extract Different Parts of data from the data column 
covid_df['year'] = pd.DatetimeIndex(covid_df.date).year
covid_df['month'] = pd.DatetimeIndex(covid_df.date).month
covid_df['day'] = pd.DatetimeIndex(covid_df.date).day
covid_df['weekday'] = pd.DatetimeIndex(covid_df.date).weekday
```

### Grouping and Aggregation  

```python
covid_month_df = covid_df.groupby('month')[['new_cases' , 'new_deaths' , 'new_tests']].sum()
```

```python
covid_month_df
```

```python
# Mean 
covid_month_mean_df = covid_df.groupby('month')[['new_cases' , 'new_deaths' , 'new_tests']].mean()
```

```python
covid_month_mean_df
```

Cumulative sum of a column as a new series (Apart from grouping, another form of aggregation is the running or cumulative sum of cases, tests, or death up to each row's date.)

```python
covid_df['total_cases'] = covid_df.new_cases.cumsum()
covid_df['total_deaths'] = covid_df.new_deaths.cumsum()
covid_df['total_tests'] = covid_df.new_tests.cumsum()
```

### Merging data from multiple sources 

```python
location_df = pd.read_csv('locationsdata.csv')
```

```python
location_df
```

```python
location_df[location_df.location == 'Italy']
```

```python
covid_df['location'] = 'Italy'
```

```python
covid_df
```

```python
 # It merged the two different df in one df.
    merged_df = covid_df.merge(location_df, on='location')
```

```python
 merged_df
```

```python
merged_df['cases_per_million'] = merged_df.total_cases*1e6 / merged_df.population 
```

```python
merged_df['deaths_per_million'] = merged_df.total_cases * 1e6 / merged_df.population
```

```python
merged_df['tests_per_million'] = merged_df.total_cases * 1e6 / merged_df.population
```

```python
merged_df
```

### Writing data back to files 

```python
result_df = merged_df[['date', 
                      'new_cases',
                      'total_cases',
                      'new_deaths',
                      'total_deaths',
                      'new_tests',
                      'total_tests',
                       'cases_per_million',
                       'deaths_per_million',
                       'tests_per_million'
                      ]]
```
